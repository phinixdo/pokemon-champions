```python
masters_top8 = ['Tapu KoKo',
                'Salamence',
                'Snorlax',
                'Icineroar',
                'Gastrodon',
                'Kartana',
                'Tapu Bulu',
                'Incineroar',
                'Gengar',
                'Latias',
                'Landorus',
                'Politoed',
                'Metagross',
                'Mimikyu',
                'Snorlax',
                'Landorus',
                'Incineroar',
                'Tapu Koko',
                'Tapu Koko',
                'Gothitelle',
                'Salamence',
                'Snorlax',
                'Incineroar',
                'Amoongus',
                'Kartana',
                'Tapu Fini',
                'Tyranitar',
                'Gengar',
                'Incineroar',
                'Landorus'
                'Tapu Koko',
                'Tapu Fini',
                'Cresselia',
                'Heatran',
                'Landorus',
                'Kangaskhan',
                'Charizard',
                'Landorus',
                'Kartana',
                'Tapu Lele',
                'Cresselia',
                'Snorlax',
                'Persian',
                'Tapu Bulu',
                'Tapu Fini',
                'Metagross',
                'Landorus',
                'Volcarona']

masters_top8 = set(masters_top8)
```

```python
df['Master Team'] = np.where(df['Name'].isin(masters_top8), 1, 0)
df[df['Name'].isin(masters_top8)].head()
```

```python
g = sns.FacetGrid(data=df, hue='Master Team', size=5)
g.map(sns.kdeplot, 'hp');
```

```python
data = pd.melt(df, id_vars=['Name','Master Team'])

g = sns.FacetGrid(data=data, hue='Master Team', size=5, col='variable')
g.map(sns.kdeplot, 'value', shade=True);
```

All stats but attack are higher, which is what we would expect. 

For further analysis, we can narrow the general sample to only include highest evolved forms from each species line.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df.drop(['Master Team', 'Name'], axis=1), df['Master Team'])

clf = LogisticRegression()

clf.fit(X_train, y_train)
pred = clf.predict(X_test)
clf.score(X_test, y_test)
```
